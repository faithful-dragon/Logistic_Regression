```python
X_train , X_test, Y_train , Y_test = train_test_split(x, y, test_size =0.2 , random_state = 42)
```

```python
# Intialising weights and biases
w =  torch.rand((x.size()[1],1),requires_grad=True)
b = torch.rand(1,requires_grad=True)
w.shape,b.shape
```

```python
#sigmoid function
def sigmoid(x):
    """Sigmoid function that squashes the input between 0 and 1"""
    activation = torch.reciprocal(torch.add(1,torch.exp(torch.neg(x))))
    return activation
```

```python
# logistic model
def model(x):
    #linear regression x * w^T + b
    linear_ =  torch.add(torch.matmul(x,w),b)
    # bound the output with sigmoid activation
    linear_activation =  sigmoid(linear_)
    return linear_activation
```

```python
def bce_loss(y_true, y_pred):
    """binary_cross_entropy
    Loss function for the training of the logistic regression"""
    # lambda or 1/N in above figure
    lambda_ =  1/(y_true.size()[0])
    # epsilon is added to contents in log function, this is to prevent finding log of zero as sigmoid is always greater or equal to zero
    epsilon = 1e-10

    # calculate
    m1 = torch.mul(y_true,torch.log(torch.add(y_pred,epsilon)))
    m2 = torch.mul(torch.add(1,torch.neg(y_true)),torch.log(torch.add(1+epsilon,torch.neg(y_pred))))
    lm = torch.add(m1,m2)
    l_theta = torch.mul(torch.sum(lm),-lambda_)
    return l_theta
```

```python
# Adjust weights & reset gradients
# Train for 10000 epochs

for i in range(10000):
    Y_pred =  model(X_train)
    loss =  bce_loss(Y_train,Y_pred)

    # calculate gradient w.r.t loss (backprop)
    loss.backward(retain_graph=True)
    w_grad = w.grad
    b_grad = b.grad
    learning_rate = 0.005
    with torch.no_grad():
        w -= torch.mul(learning_rate,w_grad)
        b -= torch.mul(learning_rate,b_grad)
        w.grad.zero_()
        b.grad.zero_()

    if i%1000 == 0:
      print(f"epoch {i} loss {loss.item()}")
```
